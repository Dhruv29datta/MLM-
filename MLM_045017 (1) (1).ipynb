```python
def count_alphanumeric_variables(df_filtered_final):
    count = 0
    for col in df_filtered_final.columns:
        if df_filtered_final[col].astype(str).str.isalnum().all():
            count += 1
    return count

num_alphanumeric_variables = count_alphanumeric_variables(df_filtered_final)

print("Number of variables containing alphanumeric values:", num_alphanumeric_variables)
```

#### Report non categorical data outliers using box plot . if yes then remove them out by either normalize (preferred) or standardize  or log transformation.

```python
non_categorical_columns = df_encoded.select_dtypes(exclude=['object']).columns

plt.figure(figsize=(100, 50))
for i, col in enumerate(non_categorical_columns, 1):
    plt.subplot(1, len(non_categorical_columns), i)
    df_encoded.boxplot(col)
    plt.title(col)
plt.tight_layout()
plt.show()



def remove_outliers(df_encoded, columns): # Removing outliers using normalization
    for col in columns:
        df_encoded[col] = (df_encoded[col] - df_encoded[col].mean()) / df_encoded[col].std()  # Normalizing the data
        df_encoded = df_encoded[abs(df_encoded[col]) <= 3] # Removing outliers (values outside 3 standard deviations from the mean)
    return df_encoded


df_cleaned = remove_outliers(df_encoded.copy(), non_categorical_columns)
plt.figure(figsize=(100, 50))
for i, col in enumerate(non_categorical_columns, 1):
    plt.subplot(1, len(non_categorical_columns), i)
    df_cleaned.boxplot(col)
    plt.title(col)
plt.tight_layout()
plt.show()
```

```python
from scipy.stats import f_oneway

# Assuming 'df' is your DataFrame
# Assuming 'kmeans' is your KMeans model with k=3

# Joining cluster labels with the original dataset
df_with_clusters = df.copy()
df_with_clusters['Cluster_Label'] = kmeans.labels_

# Extracting non-categorical variables relevant to your dataset
non_cat_variables = ['hour', 'day', 'reach', 'retweetcount', 'likes', 'klout', 'sentiment']

# Grouping variables by cluster label
cluster_groups = df_with_clusters.groupby('Cluster_Label')

# Perform ANOVA for each non-categorical variable
anova_results_non_cat = {}
for column in non_cat_variables:
    # Collecting groups for the ANOVA test
    groups = [group[column].dropna() for name, group in cluster_groups]
    # Ensure there are at least two groups with non-null values to compare
    if len(groups) > 1 and all(group.size > 0 for group in groups):
        anova_result = f_oneway(*groups)
        anova_results_non_cat[column] = anova_result

# Print ANOVA results for non-categorical variables
for column, result in anova_results_non_cat.items():
    print(f"Variable: {column}")
    print(f"F-value: {result.statistic}")
    print(f"P-value: {result.pvalue}")
    print()
```

```python
from scipy.stats import chi2_contingency
import pandas as pd

# Assuming 'df' is your DataFrame, and it now includes a 'Cluster_Label' column from KMeans clustering
# Let's first add the cluster labels to your DataFrame if you haven't done so
# Assuming 'kmeans' is your KMeans model object after fitting

df['Cluster_Label'] = kmeans.labels_

# Extracting categorical variables from your dataset
cat_variables = ['weekday', 'lang', 'isreshare']  # Identified as potential categorical variables

# Perform Chi-square test for each categorical variable
chi2_results_cat = {}
for column in cat_variables:
    if column in df.columns:
        # Creating a contingency table between each categorical variable and the cluster labels
        contingency_table = pd.crosstab(df[column], df['Cluster_Label'])
        # Performing the Chi-square test
        chi2, p_value, _, _ = chi2_contingency(contingency_table)
        # Storing the results
        chi2_results_cat[column] = {'Chi-square': chi2, 'P-value': p_value}

# Print Chi-square test results for categorical variables
for column, result in chi2_results_cat.items():
    print(f"Variable: {column}")
    print(f"Chi-square: {result['Chi-square']}")
    print(f"P-value: {result['P-value']}")
    print()
```

```python
import pandas as pd
from scipy.stats import f_oneway
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame that includes non-categorical data and 'Cluster_Label'

# Define non-categorical variables
non_cat_variables = ['hour', 'day', 'reach', 'retweetcount', 'likes', 'klout', 'sentiment']

# Initialize a dictionary to hold p-values
anova_p_values = {}

# Group the data by cluster label
cluster_groups = df.groupby('Cluster_Label')

# Perform ANOVA and collect p-values
for var in non_cat_variables:
    # Collecting groups for the ANOVA test
    groups = [group[var].dropna() for _, group in cluster_groups]
    if len(groups) > 1:  # Ensuring there are at least two groups to compare
        _, p_value = f_oneway(*groups)
        anova_p_values[var] = p_value

# Convert the p-values into a DataFrame
p_values_df = pd.DataFrame(list(anova_p_values.items()), columns=['Variable', 'P-Value']).set_index('Variable')

# Generate a heatmap for the p-values
plt.figure(figsize=(8, 4))
sns.heatmap(p_values_df.T, annot=True, cmap='viridis', cbar_kws={'label': 'P-Value'})
plt.title('ANOVA P-Values Heatmap for Non-Categorical Variables Across 3 Clusters')  # Updated title
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Assuming 'df' is your DataFrame
# Selecting and standardizing the data for clustering
features = df[['retweetcount', 'likes']]
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# Perform K-means clustering for k=3
kmeans = KMeans(n_clusters=3, random_state=111)
kmeans.fit(features_scaled)

# Add the cluster labels to your DataFrame
df['Cluster_Label'] = kmeans.labels_

# Plot the scatter plot with clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='retweetcount', y='likes', hue='Cluster_Label', legend='full', palette='viridis')

# Plot centroids
# Note: Centroid coordinates need to be inversely transformed if original data was scaled
centroids = scaler.inverse_transform(kmeans.cluster_centers_)
plt.scatter(centroids[:, 0], centroids[:, 1], s=100, c='red', label='Centroids')

# Set plot title and labels
plt.title('Clusters with Centroids for k=3')
plt.xlabel('Retweet Count')
plt.ylabel('Likes')
plt.legend()
plt.grid(True)

plt.show()

```
